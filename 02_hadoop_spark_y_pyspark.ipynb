# Introducción.

## *Big Data*.

Son las técnicas y herramientas utilizadas para procesar datos que por su cantidad o por su falta de estructura no pueden ser procesados por una sola computadora para satifacer las 4 "V".

* Volumen
* Velocidad
* Variedad
* Veracidad

## El ecosistema de *Hadoop*.

## *Data Lakehouse*. 

Es un repositorio de almacenamiento centralizado que contienen grandes cantidades de datos provenientes de múltipes fuentes en formato *crudo* y generalmente, no estructurado.  

## *Apache Spark*.

Es un framework de computación (entorno de trabajo) en clúster de ccódigo abierto. Fue desarrollada originariamente en la Universidad de California y fue donado más tarde a la *Apache Software Foundation* que se encarga de su mantenimiento desde entonces. *Spark* proporciona una interfaz para la programación de clusters completos con paralelismo de datos implícito y tolerancia a fallos.

*Apache Spark* se puede considerar un sistema de computación en clúster de propósito general y orientado a la velocidad. Proporciona *API*s en Java, Scala, Python y R. También proporciona un motor optimizado que soporta la ejecución de graficos en general. También soporta un conjunto extenso y rico de herramientas de alto nivel entre las que se incluyen *Spark SQL* (para el procesamiento de datos estructurados basada en SQL), *MLlib* para implementar machine learning, GraphX para el procesamiento de graficos y *Spark Streaming*.
